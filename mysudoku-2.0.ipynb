```python
import numpy as np 

mysudoku = np.array([
            [3,0,6,5,0,8,4,0,0]
           ,[5,2,0,0,0,0,0,0,0]
           ,[0,8,7,0,0,0,0,3,1]
           ,[0,0,3,0,1,0,0,8,0]
           ,[9,0,0,8,6,3,0,0,5]
           ,[0,5,0,0,9,0,6,0,0]
           ,[1,3,0,0,0,0,2,5,0]
           ,[0,0,0,0,0,0,0,7,4]
           ,[0,0,5,2,0,6,3,0,0]])

calllist = np.array([[i,j] for i in range(0,9) for j in range(0,9)])

def findPossible(x,y):
    onetonineset = set(range(1,10))
    #print (onetonineset)
    row = set(mysudoku[x,])
    col = set(mysudoku[:,y])
    squarex = x//3
    squarey = y//3
    #print ("square = ", squarex, squarey)
    #print ("row: ", row)
    #print ("col: ", col)
    rowset = onetonineset.difference(row)
    colset = onetonineset.difference(col)
    finalset = rowset.intersection(colset)
    square = set((mysudoku[squarex*3:squarex*3+3,squarey*3:squarey*3+3]).reshape(9))
    #print ("squareset: " , square)
    finalset = finalset.difference(square)
    out = list(finalset)
    #print ("finalset: ", finalset)
    return out

def getNext(x,y):
    (a,b) = (x,y+1) if y < 8 else (x+1,0)
    return (a,b) if mysudoku[a,b] == 0 else getNext(a,b)

def getPrev(x,y):
    (a,b) = (x-1,8) if y == 0 else (x,y-1)
    return (a,b) if mysudoku[a,b] == 0 else getPrev(a,b)

def isValid(x,y,value):
    return (True if set(findPossible(x,y)).intersection(set([value])) else False)


# use this data struct for mysudoku program
trials_dict = {}
#test_dict = {(1,0):[[1,2,3],[1,2]]}
#print(test_dict)
#print(test_dict[(1,0)][1])
#positions_left = list(set(test_dict[(1,0)][0]).difference(set(test_dict[(1,0)][1])))
#print(positions_left[0])
#test_dict[(1,0)][1].append(positions_left[0])
#print(test_dict[(1,0)][1])
#positions_left = list(set(test_dict[(1,0)][0]).difference(set(test_dict[(1,0)][1])))
#print(positions_left[0]) if positions_left else print("None")


# fillstack at positions there are zeros 
print ('i am here')
empty_cells = []
for x,y in calllist: 
    if mysudoku[x,y] == 0:
        possiblelist = findPossible(x,y)
        #print("possibleset: " , possiblelist)
        trials_dict[(x,y)] = {'possiblelist': possiblelist,'traversed': []}
        empty_cells.append((x,y))

print ('empty cells: ' , empty_cells)
print(trials_dict)
print ("isValid: " , isValid(0,1,7))

# start infinite loop. 
# 1 get next position
# 2 check if it is valid ? (intersect with row, column, square should return None)  
# 3 if it is valid then    
#     set mysudoku (x,y) to this value  
#     append to trials_dict traversed values  
#     repeat from step 1 
# 4 if not valid then check with next element and repeat from 2  
# 5 if no valid moves found
#     empty traversed values of dictionary 
#     goPrevious. 
#     try next element repeat from step 2 
index = 0
while 1==1:
    (x,y) = empty_cells[index]
    #print ('at cell', (x,y))
    element = trials_dict[(x,y)]
    #print ('element', element) 
    possible = list(set(element['possiblelist']).difference(set(element['traversed'])))
    #print('possible', possible)
    found = False
    for k in range(0,len(possible)):
        print (possible[k])
        print(isValid(x,y,possible[k]))
        if isValid(x,y,possible[k]):
            mysudoku[x,y] = possible[k]
            trials_dict[(x,y)]['traversed'].append(possible[k])
            print ('trials_dict[(x,y)]', trials_dict[(x,y)])     
            found = True
            index = index +1
            break
    if found:
        None
    else: 
        print ('not found')
        trials_dict[(x,y)]['traversed'] = []
        index = index - 1 
        mysudoku[empty_cells[index]] = 0
    print('no of empty cells', len(empty_cells))
    print('index', index)
    if index == len(empty_cells)-1:
        break
print (mysudoku)

```
