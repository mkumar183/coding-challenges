```python
# One way to unlock an Android phone is through a pattern of swipes across a 1-9 keypad.
# For a pattern to be valid, it must satisfy the following:
# All of its keys must be distinct.
# It must not connect two keys by jumping over a third key, unless that key has already been used.
# For example, 4 - 2 - 1 - 7 is a valid pattern, whereas 2 - 1 - 7 is not.
# Find the total number of valid unlock patterns of length N, where 1 <= N <= 9.
def getAdjacent (x):
    if x == 0:
        return ([7,8,9]);
    elif x == 1:
        return ([2,4,5]);
    elif x == 2:
        return ([1,3,4,5,6]);
    elif x == 3:
        return ([2,5,6]);
    elif x == 4:
        return ([1,2,5,7,8]);
    elif x == 5:
        return ([1,2,3,4,6,7,8,9]);
    elif x == 6:
        return ([2,3,5,8,9]);
    elif x == 7:
        return ([4,5,8,0]);
    elif x == 8:
        return ([4,5,6,7,9,0]);
    elif x == 9:
        return ([5,6,8,0]);


# logic that can be used is patterns of n numbers will be 0-9 and number of patterns n-1
# when n = 1 : return number of possibilities from getAdjacent 
# if any of the adjacent node is used, skip the node and get nodes adjacent to that
def emitPatterns(pre,n):
    # first digit can be any number 
    if n == 0:
        print(pre)
        return
    if len(pre) == 0:
        for i in range(0,10):
            emitPatterns([i],n-1)
    else: 
        adj = getAdjacent(pre[-1])
        for adjelem in adj:
            if adjelem not in pre:
                emitPatterns(pre + [adjelem],n-1)
                
pre = []
emitPatterns(pre,4)

#todo: have to enhance to skip the repeated digit and allow subsequent digits
```
